```python
import numpy as np
import pandas as pd

#a one-dimensional labeled array holding data of any type such as integers, strings, Python objects etc.

#a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

df = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=["A","B","C"])

print(df.head())#shows first 5 rows but since we only have 3 it will show only 3

print(df.columns) #prints the columns and here is the eact output: Index(['A', 'B', 'C'], dtype='object')

print(df) #prints the whole thing

#the first 3 numbers you see(0,1,2) those are called the index

print(df.index) #prints out the index

print(df.index.tolist())#to make it more readable and better

df1 = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=["A","B","C"], index=["x","y","z"])

print(df1) #now we have specified our index and it can be changed

print(df1.index)#you can see that is shows xyz

print(df.info())
print(df.info())
#shows you information about your dataframe

#we can see we have 3 different columns are are all 1nt64 type

#int64 = 64 bits = 8bytes

#best way to be efficient is to change these datatypes

print(df.describe())
print(df1.describe())
#tell us meaningful information about our data
#tell us things like number of items in each column
#mean of items in a column
#standard deviaton, minimum, etc

print(df.nunique()) #helps you find unique values in each column

print(df['A'].unique())#you can choose specific column see the unqie values in that column

print(df.shape)#tells us the shape of the dataframe like 2,2 or 3,3 in this case
```

```python
#loading in data
coffee = pd.read_csv("coffee.csv")#we downloaded coffee.csv data and is ina  file then loaded it in here

#csv might be the most popular but mght not be the best for your use case

print(coffee.head())

results = pd.read_parquet("results (1).parquet")
print(results.head()) # we just loaded in a datafile with bunch of olympic results

olympics_data = pd.read_excel("olympics-data.xlsx", sheet_name="results")#by default it will give us the same results as the lines above but we can specify what sheet to get different data output
print(olympics_data.head())

bios = pd.read_csv("bios.csv")
print(bios.head())

#accesing data using pandas
print(coffee)#lets you see the whole thing

print(coffee.head(10))#lets you see the first 10 rows

#.head() is a method that returns the first n rows of a DataFrame.

print(coffee.tail())#returns the last 5 rows

print(coffee.tail(10))#returns last 10 rows

print(coffee.sample())#will let you see a random sample of the data and will return 1 row as default

print(coffee.sample(10))#will return 10 random rows from the data
```

```python
#loc and iloc

#loc allows us to filter our dataframe by rows and columns
#print(coffee.loc[#rows, #columns])

print(coffee.loc[0])#will give us the first row

print(coffee.loc[[0,1,2]])#gives is the first, second, and third row

#those numbers are based on the indexes of the data

print(coffee.loc[[0,6,7]])

#you can also use slice sytnax

print(coffee.loc[0:3]) #prints rows from 0 to 3

print(coffee.loc[5:])#prints all rows from 5 to the end

print(coffee.loc[5:12]) #prints rows from 5 to 12

#now we can also use the columns
print(coffee.loc[5,"Day"])#prints index 5 but just shows the day

print(coffee.loc[5:12,"Day"])#shows only day for rows 12-5

print(coffee.loc[5:12,["Day","Units Sold"]]) # shows us only the day and units sold column for rows 5-12

print(coffee.loc[:,["Units Sold", "Day"]])#shows us all the rows and only the day and units sold columns

print(coffee.iloc[:,[0,2]]) #gives us the same result as the code above

#basically when we use iloc we use index positions instead of labels

#print(coffee.iloc[:,["Day","Units Sold"]]) will give an error

print(coffee.iloc[5:8,[0,2]]) #shows columns 0-2 and rows 5-8

#the only difference in iloc is that the upper idnex is not inlcuive but with loc it is inclusive, for example is 5:8 it will only show rows 5,6,7

coffee.index = coffee.Day # this allows us to access specific columns
#coffee.index = coffee["Day"] does the same thing as the code above

print(coffee.head())#now you can see for the column names there is no number and the index is replaces with the days

#now id we try print(coffee.loc[0]) it will give us an error because we have changed the index to days

print(coffee.loc["Monday"])#now we can see the first row

print(coffee.loc["Monday":"Wednesday"])

print(coffee.loc["Monday":"Wednesday", "Units Sold"])
```

```python
#optimized ways

#you can use things like iat and more to get specific values

#coffee.at[0:3, "Units Sold"] #will give error because it can't give more than 1 value

coffee.iat[0,0]#will give 1 value and is called integer at

#coffee.iat[0:3, 0] #will give error

coffee.iloc[0,0]#does the same thing is most commonly used

#if you need to grab 1 value really efficiently then use iat and at

coffee.at[0, "Units Sold"]#gives you only 1 value
```

```python
#Sorting data
coffee.sort_values("Units Sold")#gives us the whole thing but its ordered so everything in the Unnits sold sections is in asceding order

coffee.sort_values("Units Sold", ascending = False)#Goes from greateast to least

coffee.sort_values(["Units Sold", "Coffee Type"], ascending = False)#this will first sort by units sold and if there a vlue that has the same units sold then it will sort by coffee type and will be in alphabetical order by default

coffee.sort_values(["Units Sold", "Coffee Type"], ascending=[0,1])#so units sold is sorted in greatest to least order and cofee type is sorted in alphabetical order

#The code below is not reccomended for many use cases, this is what it does: you need to iterate row by row in your dataframe and loses you memory benefits of python pandas  

for index, row in coffee.iterrows():
    print(index)
    print(row) #you could also do something like row['Units Sold']
    print("\n\n\n")

#the code above is not that great so try to stick to the pandas built in functions when you can
```

```python
#Filtering Data

bios.head()

bios.tail()

#we can filter by height and weight

bios.info()
#now we can see that both heigh and weight are float64

bios.loc[bios['height_cm'] > 215] #We are looking for all athletes whose height is bigger than 215 cm

bios.loc[bios['height_cm'] > 215,['name', 'height_cm', 'NOC']] #we can use this to make it a little bit simpler

#basically we used loc then set a conditional and at the end specified the columns we wanted to see

bios[bios['height_cm'] > 215] #shorter syntax and does the same thing

bios[bios['height_cm'] > 215] [['name', 'height_cm']] # this also does the same thing except in different format

#what if we wanted to have multipel conditions?

bios[(bios['height_cm'] > 215) & (bios['born_country'] == 'USA')] #this gives us all the athlets from the usa and are taller than 215 cm
```

```python
#more interesting and specific filtering methods

bios[bios['name'].str.contains('Keith')]#ths is case sensitive so you have to put Keith with capital K

bios[bios['name'].str.contains('keith', case=False)]#here no you can type keith with lowercase k

bios[bios['name'].str.contains('keith|patrick', case=False)]#this basically searches for people naemd keith OR patrick hte | symbol means OR in regex format

#addtioanlal examples could be to find all the athletes who are from countries that start with an vowel
#Can find athletes with names who have exactly 2 vowels
#these thigns are extreemly powerful and extremely useful

#if you dont want to search using regex just do this
bios[bios['name'].str.contains('keith|patrick', case=False, regex=False)]#returns nothing

#other methods
bios[bios['born_country'].isin(["USA", "FRA", "GBR"])]#shows only the countries with abbreviations specified

bios[bios['born_country'].isin(["USA", "FRA", "GBR"]) & (bios["name"].str.startswith("Keith"))] #this will show everyone wiht the first name Keith and we did this on purpose by using startswith and check if they are from those 3 countries

bios.query('born_country == "USA"')#query functions are also another way to filter data 

bios.query('born_country == "USA" and born_city == "Seattle"')#you can pass in mutiple functions so this looks for the born country as usa and the city as seattle
```
